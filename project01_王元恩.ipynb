```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab-Notebooks/aqf_p_01.csv")
df.head(5)
```

```python
# 先查看個欄位資料屬性
df.dtypes
```

```python
df.describe() #連續變數的描述統計
```

```python
# 看一下 LOCATION的類別
area_count= df["area"].value_counts()
print(area_count)
```

```python
# 限制 "北部","中部","花東","高屏" 等地區做資料分佈
filt = (df["area"] == "北部")  # FILTER  
df["forecastdate"].value_counts().sort_index(ascending=True)
loc1=[]
loc1= (df.loc[filt,["area","forecastdate","aqi"]])
loc1["forecastdate"].value_counts().sort_index(ascending=True)
# print(loc1)

loc1.describe()

plt.figure(figsize=(12,6))
loc1["aqi"].hist(bins=20)
plt.xlabel("forecastdate")
plt.ylabel("aqi")
plt.title(f"空氣品質")
plt.show();

```

```python
# print(loc1) 北區Daily空氣品質

loc1_new = loc1.drop_duplicates()
# print(loc1_new)

line_plot = loc1_new["forecastdate"].value_counts().sort_index(ascending=True).plot()
plt.title("Air Quality index of Northern Taiwan")
plt.xlabel("DATE")
plt.ylabel("AQI")
plt.show()
```

```python
# 中部 DATA SET
filt = (df["area"] == "中部")  # FILTER  
df["forecastdate"].value_counts().sort_index(ascending=True)
loc2=[]
loc2= (df.loc[filt,["area","forecastdate","aqi"]])
#loc2["forecastdate"].value_counts().sort_index(ascending=True)
loc2_new = loc2.drop_duplicates()
# print(loc2_new)


# 高屏 DATA SET
filt = (df["area"] == "高屏")  # FILTER  
df["forecastdate"].value_counts().sort_index(ascending=True)
loc3=[]
loc3= (df.loc[filt,["area","forecastdate","aqi"]])
#loc3["forecastdate"].value_counts().sort_index(ascending=True)
loc3_new = loc3.drop_duplicates()
# print(loc3_new)

# 花東 DATA SET
filt = (df["area"] == "花東")  # FILTER  
df["forecastdate"].value_counts().sort_index(ascending=True)
loc4=[]
loc4= (df.loc[filt,["area","forecastdate","aqi"]])
#loc4["forecastdate"].value_counts().sort_index(ascending=True)
loc4_new = loc4.drop_duplicates() #去重覆
# print(loc4_new)
```

```python

line_plot1 = loc1_new["forecastdate"].value_counts().sort_index(ascending=True).plot(marker="X",color = "blue", linestyle = "--")
plt.title("Air Quality index of Northern Taiwan", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("DATE", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.ylabel("AQI", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.show()

line_plot2 = loc2_new["forecastdate"].value_counts().sort_index(ascending=True).plot(marker="d",color = "red", linestyle = "-.")
plt.title("Air Quality index of Central Taiwan", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("DATE", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.ylabel("AQI", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.show()

line_plot3 = loc3_new["forecastdate"].value_counts().sort_index(ascending=True).plot(marker="o",color = "green", linestyle = ":")
plt.title("Air Quality index of Southern Taiwan", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("DATE", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.ylabel("AQI", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.show()

line_plot4 = loc4_new["forecastdate"].value_counts().sort_index(ascending=True).plot(marker="v",color = "orange")
plt.title("Air Quality index of Eastern Taiwan", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("DATE", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.ylabel("AQI", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.show()

```

```python
# 並排呈現北中南東 四區空氣品質
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(loc1_new["forecastdate"].value_counts().sort_index(ascending=True))
axs[0, 0].set_title("Nor")
axs[0, 1].plot(loc2_new["forecastdate"].value_counts().sort_index(ascending=True), 'tab:orange')
axs[0, 1].set_title("Mid")
axs[1, 0].plot(loc3_new["forecastdate"].value_counts().sort_index(ascending=True), 'tab:green')
axs[1, 0].set_title("Sou")
axs[1, 1].plot(loc4_new["forecastdate"].value_counts().sort_index(ascending=True), 'tab:red')
axs[1, 1].set_title("Est")
```

```python
#設定xy軸，取北區&中區aqi資料做 散佈圖
x = loc1_new.aqi[0:21] 
y = loc2_new.aqi[0:21] 
#設定大小跟x軸兩端的值
plt.figure(figsize = (20,10))
plt.xlim(40, 80)
#做圖
plt.scatter(x, y, alpha=0.6, label = "AQI") 
plt.legend(scatterpoints=1, markerscale=2)
plt.title("Scatter plot for AQI of Northern & Central Taiwan", fontsize = 30)
plt.xlabel("AQI of Northern Taiwan")
plt.ylabel("AQI of Central Taiwan")
plt.show()
```
